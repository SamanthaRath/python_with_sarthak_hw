ELLIPSE PROBLEM


```python
import random
xlst = []
ylst = []
for i in range(0,10):
  x = random.randint(0,20)
  y = random.randint(0,20)
  xlst.append(x)
  ylst.append(y)
```

```python
def checkrange(a,b):
  for i in range(10):
    x,y = xlst[i], ylst[i]
    s = (x**2)/(a**2)+(y**2)/(b**2)-1
    if(s>0):
      print("It is lying outside the ellipse")
    if(s==0):
      print("It is lying on the ellipse")
    if(s<0):
      print("It is lying inside the ellipse")
```

```python
a=int(input("Enter half major axis"))
b=int(input("Enter half minor axis"))
r=2
i=0
while(i<20):
  checkrange((r**i)*a,(r**i)*b)
  print("_______________________")
  i+=1
```
